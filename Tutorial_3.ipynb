```python
import numpy as np
# 1. Create a numpy array of size 2x2
array_2x2 = np.array([[1,2],[3,4]])
print(array_2x2)
```

```python
# 2. Create a random numpy array of size 3x4
random_3x4 = np.random.rand(3,4)
print(random_3x4)
```

```python
# 3. Create a numpy array of ones and a numpy array of zeros
ones_array = np.ones((2,3))
zeros_array = np.zeros((2,3))
print(ones_array)
print()
print(zeros_array)
```

```python
# 4. Create a random numpy array of size 6x3 and transpose it
random_6x3 = np.random.rand(6,3)
transpose_array = random_6x3.T
print(random_6x3)
print()
print(transpose_array)
```

```python
# 5. Create two random numpy array of size n x p and p x m and then compute the dot product
n, p, m = 3, 2, 4
A = np.random.rand(n,p)
B = np.random.rand(p,m)
dot_product = np.dot(A,B)
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("dot product :\n", dot_product)
```
